## 7. Quantidade de mensagens por estado;

```python
# não existem tabelas sobre estado nesse dataset
estados = [
    "Acre", "Alagoas", "Amapá", "Amazonas", "Bahia", "Ceará", "Distrito Federal",
    "Espírito Santo", "Goiás", "Maranhão", "Mato Grosso", "Mato Grosso do Sul",
    "Minas Gerais", "Pará", "Paraíba", "Paraná", "Pernambuco", "Piauí", "Rio de Janeiro",
    "Rio Grande do Norte", "Rio Grande do Sul", "Rondônia", "Roraima", "Santa Catarina",
    "São Paulo", "Sergipe", "Tocantins"
]

sql_parts = [
    f"SELECT '{estado}' AS estado, COUNT(*) AS total FROM data_filtered WHERE LOWER(text_content_anonymous) LIKE '%{estado.lower()}%'" 
    for estado in estados
]

query = " UNION ALL ".join(sql_parts)

messages_per_states = conn.execute(query).fetchdf()

print(messages_per_states.sort_values('total', ascending=False).reset_index(drop=True))
```

## 9. Relação quantidade de usuários por quantidade de mensagens por estado;

```python
sql_parts = [
    f"""
    SELECT 
        '{estado}' AS estado,
        COUNT(*) AS total_mensagens,
        COUNT(DISTINCT id_member_anonymous) AS total_usuarios,
        CAST(COUNT(DISTINCT id_member_anonymous) AS DOUBLE) / COUNT(*) AS relacao_usuarios_por_mensagem
    FROM data_filtered
    WHERE LOWER(text_content_anonymous) LIKE '%{estado.lower()}%'
    """
    for estado in estados
]

query_relacao = " UNION ALL ".join(sql_parts)

relacao_usuarios_mensagens = conn.execute(query_relacao).fetchdf()

df_ = relacao_usuarios_mensagens.sort_values('relacao_usuarios_por_mensagem', ascending=False).reset_index(drop=True)
```

## 10. Quantidade de mensagens por país;

```python
from langdetect import detect, DetectorFactory
DetectorFactory.seed = 42  

lingua_para_pais = {
    'pt': 'Brasil',
    'es': 'Espanha / América Latina',
    'en': 'Estados Unidos / Reino Unido',
    'fr': 'França',
    'de': 'Alemanha',
    'it': 'Itália',
    'ru': 'Rússia',
    'zh-cn': 'China',
    'ar': 'Países Árabes',
    'hi': 'Índia',
}
```

```python
from collections import Counter

from tqdm import tqdm

contador_paises = Counter()

txt = df['text_content_anonymous'].dropna()

for texto in tqdm(txt):
    try:
        lang = detect(texto)
        pais = lingua_para_pais.get(lang, 'Outro')
        contador_paises[pais] += 1
    except:
        contador_paises['Indefinido'] += 1

import pandas as pd

df_paises = pd.DataFrame(contador_paises.items(), columns=['pais', 'total_mensagens']).sort_values('total_mensagens', ascending=False)
print(df_paises)
```

## 11. Quantidade de mensagens Brasil X Países Estrangeiros;

```python
def categorizar_origem(pais):
    if pais == 'Brasil':
        return 'Brasil'
    elif pais == 'Indefinido':
        return 'Indefinido'
    else:
        return 'Estrangeiro'

df_paises['origem'] = df_paises['pais'].apply(categorizar_origem)

df_origem = df_paises.groupby('origem')['total_mensagens'].sum().reset_index()
print(df_origem)
```

## 21. Os 30 unigramas, bigramas e trigramas mais compartilhados;

```python
!pip install -q nltk
```

```python
from nltk.util import ngrams
from collections import Counter
import re

# tokenizer using regex
def simple_tokenize(text):
    text = text.lower()
    text = re.sub(r'https?://\S+', '', text)  # remove URLs
    tokens = re.findall(r'\b\w+\b', text)  # catch words
    return tokens

df_texts = conn.execute("""
    SELECT text_content_anonymous
    FROM data_filtered
    WHERE text_content_anonymous IS NOT NULL
""").fetchdf()

unigram_counter = Counter()
bigram_counter = Counter()
trigram_counter = Counter()

for text in df_texts['text_content_anonymous']:
    tokens = simple_tokenize(text)
    unigram_counter.update(tokens)
    bigram_counter.update([' '.join(bg) for bg in ngrams(tokens, 2)])
    trigram_counter.update([' '.join(tg) for tg in ngrams(tokens, 3)])

print("\nTop 30 Unigrams:")
print(unigram_counter.most_common(30))

print("\nTop 30 Bigrams:")
print(bigram_counter.most_common(30))

print("\nTop 30 Trigrams:")
print(trigram_counter.most_common(30))
```
