## f. Identificar e listar as posições (células) contendo valores que não pertencem ao domínio (tipo de dados) esperado.

```python
def identify_inconsistent_cells(df):
    inconsistent_cells = {}
    
    for col_idx, column in enumerate(df.columns):
        if df[column].isna().all():
            continue
            
        types = [type(x).__name__ for x in df[column].dropna()]
        type_counts = Counter(types)
        dominant_type = type_counts.most_common(1)[0][0]
        
        for row_idx, value in enumerate(df[column]):
            if pd.isna(value):
                continue
                
            if type(value).__name__ != dominant_type:
                cell_position = (row_idx, col_idx)
                column_name = column
                
                inconsistent_cells[cell_position] = {
                    'column_name': column_name,
                    'value': value,
                    'actual_type': type(value).__name__,
                    'expected_type': dominant_type
                }
    
    return inconsistent_cells

results = identify_inconsistent_cells(df)

print(f"Found {len(results)} cells with inconsistent data types:")
for (row, col), info in results.items():
    print(f"Cell position: Row {row}, Column {col} ({info['column_name']})")
    print(f"  Value: {info['value']}")
    print(f"  Actual type: {info['actual_type']}, Expected type: {info['expected_type']}")
    print()
```

## k. Crie uma coluna chamada “sentiment” contendo os valores: -1 para textos negativos, 0 para textos neutros e 1 para textos positivos

```python
!pip -q install nltk
```

```python
import nltk 
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')
```

```python
sid = SentimentIntensityAnalyzer()

def get_sentiment_category(text):
    if not isinstance(text, str) or not text.strip():
        return 0  # neutral or empty strings
    
    scores = sid.polarity_scores(text)
    compound = scores['compound']
    
    if compound <= -0.05:
        return -1  # negative
    elif compound >= 0.05:
        return 1   # positive
    else:
        return 0   # neutral

df['sentiment'] = df['text_content_anonymous'].apply(get_sentiment_category)
```

## m. Identificar inconsistências entre os atributos (features).

```python
inconsistencies_media = df[
    (df['has_media'] == True) & (
        df['media'].isnull() | df['media_type'].isnull() | df['media_name'].isnull()
    )
]

inconsistencies_media_url = df[
    (df['has_media_url'] == True) & (df['media_url'].isnull())
]


df['date_message'] = pd.to_datetime(df['date_message'], errors='coerce')
df['date_system'] = pd.to_datetime(df['date_system'], errors='coerce')

inconsistencies_datas = df[
    df['date_message'] > df['date_system']
]

inconsistencies_score_sentiment = df[
    (df['score_sentiment'] < -1) | (df['score_sentiment'] > 1)
]

inconsistencies_score_misinformation = df[
    (df['score_misinformation'] < 0) | (df['score_misinformation'] > 1)
]

inconsistencies_ids = df[
    df['id_member_anonymous'].isnull() |
    df['id_group_anonymous'].isnull() |
    df['id_message'].isnull()
]

print(f"Inconcistencies - Media: {len(inconsistencies_media)} rows")
print(f"Inconcistencies - Media URL: {len(inconsistencies_media_url)} rows")
print(f"Inconcistencies - Datas: {len(inconsistencies_datas)} rows")
print(f"Inconcistencies - Score Sentiment: {len(inconsistencies_score_sentiment)} rows")
print(f"Inconcistencies - Score Misinformation: {len(inconsistencies_score_misinformation)} rows")
print(f"Inconcistencies - IDs: {len(inconsistencies_ids)} rows")
```
